```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
url = "https://community.watsonanalytics.com/wp-content/uploads/2015/04/WA_Fn-UseC_-Sales-Win-Loss.csv"
```

```python
df = pd.read_csv(url)
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns = [each.lower() for each in df.columns]
```

```python
sns.set(rc={'figure.figsize': (11.7, 8.27)})
```

```python
sns.countplot(x='route to market', data=df, hue='opportunity result')
sns.despine(offset=10, trim=True)
plt.show()
```

```python
sns.violinplot(x='opportunity result', y='client size by revenue', hue='opportunity result', data=df);
```

```python
print("Supplies Subgroup' : ",df['supplies subgroup'].unique())
print("Region : ",df['region'].unique())
print("Route To Market : ",df['route to market'].unique())
print("Opportunity Result : ",df['opportunity result'].unique())
print("Competitor Type : ",df['competitor type'].unique())
print("'Supplies Group : ",df['supplies group'].unique())
```

```python
## we xouls use this in the fifa project. 

from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
df['supplies subgroup'] = le.fit_transform(df['supplies subgroup'])
df['region'] = le.fit_transform(df['region'])
df['route to market'] = le.fit_transform(df['route to market'])
df['opportunity result'] = le.fit_transform(df['opportunity result'])
df['competitor type'] = le.fit_transform(df['competitor type'])
df['supplies group'] = le.fit_transform(df['supplies group'])

df.head()
```

```python
cols = [col for col in df.columns if col not in ['opportunity number, opportunity result']]
X = df[cols]

y = df['opportunity result']

df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                test_size=.3,
                                                random_state=10)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

### Gaussian NB

```python
gnb = GaussianNB()

preds = gnb.fit(X_train,y_train).predict(X_test)

print('Naive-Bayse accuracy:', accuracy_score(y_test, preds, normalize=True))
```

### LinearSVC

```python
from sklearn.svm import LinearSVC
```

```python
svc= LinearSVC(max_iter=10000,random_state=42)

pred = svc.fit(X_train,y_train).predict(X_test)

print('LinearSVC accuracy:', accuracy_score(y_test, preds, normalize=True))
```

### KneighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc = KNeighborsClassifier(n_neighbors=10)
```

```python
knc.fit(X_train, y_train)

preds = knc.predict(X_test)

print('KNeighbors accuracy score:', accuracy_score(y_test, preds))
```
